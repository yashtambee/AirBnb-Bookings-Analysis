#**3. Top 10 cheapest properties listed on airbnb & their neighbourhood ?**


````python
# top 10 cheapest properties & their neighbour hood & location
cheapest_properties = airbnb_df.loc[:,['host_name','name','neighbourhood_group','neighbourhood','price','latitude','longitude','room_type']].sort_values(by = 'price',ascending = True).reset_index().drop('index',axis = 1)
cheapest_properties.head(10)
````

````python
# filtering the cheapest_properties by ['neighbourhood','price'] to plot the bar graph
top_10_cheapest_properties = cheapest_properties.head(10)[['neighbourhood','price']]
top_10_cheapest_properties
````

````python
# visualizing the above top_10_cheapest_properties data
top_10_cheapest_properties.set_index('neighbourhood').plot.bar(color = '#2596be')
plt.rcParams['figure.figsize']=(10,5)
plt.xticks(rotation = 45)
plt.xlabel('Neighbourhood', fontsize = 14,labelpad = 17)                   
plt.ylabel('Price of Property', fontsize = 14,labelpad = 17)
plt.title('Top 10 cheapest properties', fontsize = 14, pad = 19)

# adding text on each bar
for i in range(len(top_10_cheapest_properties)):  # len of x axis
  # plt.text(x cordinate,y cordinate, label value , ha = '' , va = '')
  # ha = horizontal allignment | va = vertical allignment
  plt.text(i , top_10_cheapest_properties.iloc[i,1] , top_10_cheapest_properties.iloc[i,1] , ha = 'center', va = 'bottom')

plt.ylim([0,14])   # y axis limiter 
plt.show()
````

## Conslusion :
As per above plot, the most cheapest listings on airbnb are Hells kitchen(Manhattan), Sunset park(Brooklyn), Jamaica(Queens), Upper west side(Manhattan),Bedford-Stuyvesant(Brooklyn), East Village(Manhattan), Kips Bay(Manhattan), Upper East Side(Manhattan), Williamsburg(Brooklyn), Bushwick(Brooklyn)

All this properties are listed for 10USD which means no hosts are not ready to list his/her property less than 10USD. 10usd is the miminum threshold price point that a host can afford to list his property for

Airbnb can use this data & can use it as a filter to shortlist most cheapest listings with the neighbourhood(areas) & neighbourhood_group(loactions) on thier website to help customers to find their desired listing with ease

# **6. Which neighborhood group has maximum no of listings? And also detect, from that particular neighborhood group(location) having maximum no of listings, which neighborhood(area) has maximum no of listed properties?**

## Conclusion 1 :
So from above data it is clear that Manhattan has the maximum number of properties listed counting to 21652 which results to 44.3 % of the entire listings on the airbnb

````python
# creating a subset of rows with manhattan
manhattan_subset= airbnb_df.loc[airbnb_df['neighbourhood_group']=='Manhattan',['neighbourhood_group','neighbourhood','name']]
manhattan_subset
````

#**8. Is there any noticeable difference of traffic among different areas and what could be the reason for it?**


````python
# Filtering listings where people are visiting the most
listings_with_most_consumers = airbnb_df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
listings_with_most_consumers.sort_values(by = 'minimum_nights' ,ascending = False)
````

While considering minimum nights & price/night here consumers will basically decide a rent of the property or listings on the basis of price distribution given.

For Example, Entire home/apt : spacious studio/ loft by central park(name of listing) is offering minimum 10 nights at the cost of 80 Dollar from their consumers will definitely derive the cost for 1 night as 8 Dollar

````python
# Visulazing the listings where people are visiting the most

plt.rcParams['figure.figsize'] = (7,5)
room_type = listings_with_most_consumers['room_type']
min_night_stay = listings_with_most_consumers['minimum_nights'] 

plt.bar(room_type,min_night_stay, width = 0.6)
````

## Conclusion :
Also from the above chart we can clearly say that people are giving their preference to stay in entire home / apartment or private room present in Manhattan, Queens, and Brooklyn and also consumers are prefering the listings which are less in price
